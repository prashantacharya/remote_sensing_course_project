```python
image_paths = {
    '2018': 'dataset/2018/m_3912112_sw_10_060_20180718.tif',
    '2020': 'dataset/2020/m_3912112_sw_10_060_20200709.tif',
    '2022': 'dataset/2022/m_3912112_sw_10_060_20220714.tif'
}
```

```python
import rasterio

with rasterio.open(image_paths['2018']) as src:
    print(src.meta)  # Print metadata
    print(src.crs)
```

```python
with rasterio.open(image_paths['2020']) as src:
    print(src.meta)
    print(src.crs)
```

```python
with rasterio.open(image_paths['2018']) as src_2018:
    print(f"Number of Bands: {src_2018.count}")
    for i in range(1, src_2018.count + 1):
        print(f"Band {i} Metadata: {src_2018.tags(i)}")
```

```python
import rasterio
from rasterio.enums import Resampling

# Load the 2018 image
with rasterio.open(image_paths['2018']) as src_2018:
    red_2018 = src_2018.read(3)  # Band 3 = Red
    nir_2018 = src_2018.read(4)  # Band 4 = NIR
    transform_2018 = src_2018.transform  # Transformation info for 2018
    crs_2018 = src_2018.crs  # Coordinate reference system for 2018

# Load the 2022 image and resample it to match the 2018 dimensions
with rasterio.open(image_paths['2022']) as src_2022:
    # Resample the 2022 image to match the 2018 image shape
    data_2022_resampled = src_2022.read(
        out_shape=(src_2022.count, red_2018.shape[0], red_2018.shape[1]),
        resampling=Resampling.bilinear
    )
    
    # Extract the resampled Red and NIR bands
    red_2022 = data_2022_resampled[2]  # Assuming Band 3 (Red) is the 3rd band
    nir_2022 = data_2022_resampled[3]  # Assuming Band 4 (NIR) is the 4th band
    transform_2022 = src_2022.transform  # Transformation info for 2022
    crs_2022 = src_2022.crs  # Coordinate reference system for 2022
```

```python
import numpy as np

scaling_factor = 0.0001  # Scaling factor for NAIP imagery (to convert DNs to reflectance)

# Scale the Red and NIR bands for both years
red_2018_scaled = red_2018 * scaling_factor
nir_2018_scaled = nir_2018 * scaling_factor

red_2022_scaled = red_2022 * scaling_factor
nir_2022_scaled = nir_2022 * scaling_factor


# Calculate NDVI for 2018
ndvi_2018 = (nir_2018_scaled - red_2018_scaled) / (nir_2018_scaled + red_2018_scaled)

# Calculate NDVI for 2022
ndvi_2022 = (nir_2022_scaled - red_2022_scaled) / (nir_2022_scaled + red_2022_scaled)
```

```python
import matplotlib.pyplot as plt

# Plot NDVI for 2018
plt.figure(figsize=(10, 6))
plt.imshow(ndvi_2018, cmap='RdYlGn', vmin=-1, vmax=1)
plt.colorbar(label='NDVI')
plt.title('NDVI for 2018')
plt.show()

# Plot NDVI for 2022
plt.figure(figsize=(10, 6))
plt.imshow(ndvi_2022, cmap='RdYlGn', vmin=-1, vmax=1)
plt.colorbar(label='NDVI')
plt.title('NDVI for 2022')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ndvi_diff = ndvi_2022 - ndvi_2018

plt.figure(figsize=(10, 6))
plt.hist(ndvi_diff.flatten(), bins=50, color='blue', alpha=0.7)
plt.title('NDVI Difference Histogram')
plt.xlabel('NDVI Difference')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Alternatively, using Seaborn
sns.histplot(ndvi_diff.flatten(), bins=50, kde=False, color='blue')
plt.title('NDVI Difference Histogram')
plt.xlabel('NDVI Difference')
plt.ylabel('Frequency')
plt.show()

sns.kdeplot(ndvi_diff.flatten(), color='red')
plt.title('NDVI Difference KDE Plot')
plt.xlabel('NDVI Difference')
plt.ylabel('Density')
plt.show()

sns.histplot(ndvi_diff.flatten(), bins=50, kde=True, color='blue')
plt.title('NDVI Difference Histogram and Density')
plt.xlabel('NDVI Difference')
plt.ylabel('Frequency/Density')
plt.show()
```

```python
import numpy as np
import rasterio

# Load the RGB image (assuming it's in 'image_paths['2018']')
with rasterio.open(image_paths['2018']) as src_2018:
    # Read all 3 bands (RGB)
    red = src_2018.read(1)  # Red band
    green = src_2018.read(2)  # Green band
    blue = src_2018.read(3)  # Blue band

# Define RGB ranges for light green and dark green
# Light green: R  100-200, G  200-255, B  100-150
light_green_mask = ((red >= 100) & (red <= 200)) & \
                   ((green >= 150) & (green <= 255)) & \
                   ((blue >= 100) & (blue <= 150))

# Dark green: R  0-100, G  50-150, B  0-50
dark_green_mask = ((red >= 0) & (red <= 100)) & \
                  ((green >= 50) & (green <= 150)) & \
                  ((blue >= 0) & (blue <= 50))

# Apply the masks to create binary images of light and dark green areas
light_green_area = np.where(light_green_mask, 1, 0)  # 1 for light green, 0 for others
dark_green_area = np.where(dark_green_mask, 1, 0)  # 1 for dark green, 0 for others
```

```python
import matplotlib.pyplot as plt

# Plot the light green areas
plt.figure(figsize=(10, 6))
plt.imshow(light_green_area, cmap='Greens')
plt.title("Light Green Areas (Vegetation)")
plt.show()

# Plot the dark green areas
plt.figure(figsize=(10, 6))
plt.imshow(dark_green_area, cmap='Greens')
plt.title("Dark Green Areas (Forest)")
plt.show()
```
